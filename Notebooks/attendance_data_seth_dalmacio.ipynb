```python
import pandas as pd
import numpy as np

# Load the CSV file into a DataFrame
attendance_data = pd.read_csv("/Users/ylsstudent/Documents/Data_Analytics_Bootcamp/Projects/BootcampProject3/nfl_project/Resources/attendance_data.csv")

attendance_data.head()
```

```python
# Rename the 'Tm' column to 'team_name'
attendance_data.rename(columns = {"Tm": "team_name"}, inplace=True)

# Drop the last row of data
attendance_data = attendance_data[:-1]

attendance_data.tail()
```

```python
# Sort the DataFrame by 'team_name' in alphabetical order
attendance_data.sort_values(by = "team_name", inplace = True)

attendance_data.head()
```

```python
# Reset index to ensure unique IDs start from 1
attendance_data = attendance_data.sort_values(by = "team_name").reset_index(drop = True)

attendance_data.head()
```

```python
team_ids = []

for team_id in range(1, len(attendance_data) + 1):
    team_ids.append(team_id)

attendance_data["team_id"] = team_ids

attendance_data.head()
```

```python
attendance_data.insert(0, "team_id", attendance_data.pop("team_id"))

attendance_data.head()
```

```python
attendance_data.dtypes
```

```python
attendance_data = attendance_data.replace(",", "", regex = True)

attendance_data.head()
```

```python
attendance_data.replace("Bye", np.nan, inplace = True)

attendance_data.head()
```

```python
for col in attendance_data.columns:
    if col not in ["team_id", "team_name"]:
        attendance_data[col] = pd.to_numeric(attendance_data[col], errors = "coerce").astype("Int64")

attendance_data.dtypes
```

```python
attendance_data = attendance_data[attendance_data.columns.drop(list(attendance_data.filter(regex='Week')))]

attendance_data.head()
```

```python
attendance_data = attendance_data.rename(columns={
    "Total": "total_attendance",
    "Home": "home_attendance",
    "Away": "away_attendance"})
                                        
```

```python
attendance_data = attendance_data.drop(columns = ["team_name"])
```
